```python
import pandas as pd


animauxCsv = pd.read_csv("/Users/yassine/Desktop/data-python/fr_animaux.csv", sep=',')

DateAnmiaux = animauxCsv.loc[animauxCsv['Année'] == 2013]

synthesis = DateAnmiaux.drop(['Symbole','Domaine','Code zone','Code Domaine','Code Élément','Code année','Description du Symbole'], axis=1)
synthesis.rename(columns={'Valeur':'proteine'}, inplace=True)


indexNames = synthesis[synthesis['Élément'] == 'Disponibilité de protéines en quantité (g/personne/jour)']


res = indexNames.groupby(['Zone']).agg({'proteine': sum});



res
```
